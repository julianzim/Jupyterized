____________Решил обучить модель предсказывать цену купленного билета в зависимости от возраста_____________

```python
titanic = sns.load_dataset('titanic')
titanic = titanic.dropna(how='any', axis=0) #удалил строки с пустыми значениями
titanic
```

```python
%matplotlib inline

plt.scatter(titanic['age'], titanic['fare'])
```

```python
X, y = np.array(titanic['age'])[:, np.newaxis], np.array(titanic['fare'])[:, np.newaxis]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7) 
#создал тренировочную и тестовую выборки
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()    #создаем экземпляр модели
model.fit(X_train, y_train)   #обучаем модель
```

```python
y_pred = model.predict(X_test)   #предсказываем таргет на тестовых фичах
```

```python
model.coef_   #найденный коэффициент модели линейной регрессии
```

```python
#проверяем метрики качества

from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import r2_score

print(f'Mean squared error = {MSE(y_test, y_pred)}')
print(f'R2_Score = {r2_score(y_test, y_pred)}')
```

```python
#визуализируем тестовую выборку и предсказание

plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red', linewidth=3)
```

________________________ТЕПЕРЬ ПОПРОБУЕМ ОБУЧЕНИЕ С ПРЕОБРАЗОВАННЫМИ ДАННЫМИ_________________________

В качестве _чисто инженерных приёмов_ первичной обработки данных для борьбы с большими 
по модулю выбросами обычно используют:
    • логарифмирование np.log
    • извлечение квадратного корня np.sqrt
Оба этих преобразования являются монотонными, т.е. они меняют абсолютные значения,
но сохраняют порядок величин Пример немонотонного преобразования для величины x – функция sin(x)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
titanic = sns.load_dataset('titanic').dropna(how='any', axis=0)
titanic
```

```python
X, y = np.array(titanic['age'])[:, np.newaxis], np.array(titanic['fare'])[:, np.newaxis]
```

```python
plt.scatter(X, y)   #сырые данные. есть выбросы по целевой переменной
```

```python
y_trans = np.sqrt(y)   #если прологарифмировать то появятся бесконечные значения. поэтому взял корень
plt.scatter(X, y_trans)   #выбросы стали уже намного меньше
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as MSE

X_train, X_test, y_train, y_test = train_test_split(X, y_trans, test_size=0.2, random_state=7)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(f'Mean squared error = {MSE(y_test, y_pred)}')
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
```

Вот, уже гораздо лучше. MSE уменьшилась на 2 порядка

Теперь попробуем встроенные инструменты sklearn.preprocessing: MinMaxScaler и StandartScaler

```python
from sklearn.preprocessing import MinMaxScaler

X_transed = MinMaxScaler().fit_transform(X)
y_transed = MinMaxScaler().fit_transform(y)

plt.scatter(X_transed, y_transed)
```

----------диапазон значений изменился, но т.к. сохранился масштаб то и выброс тоже остался. нормально ли это?----------

```python
X_train, X_test, y_train, y_test = train_test_split(X_transed, y_transed, test_size=0.2, random_state=7)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(f'Mean squared error = {MSE(y_test, y_pred)}')
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
```

----------MSE стала еще меньше! 

```python
from sklearn.preprocessing import StandardScaler

X_transed = StandardScaler().fit_transform(X)
y_transed = StandardScaler().fit_transform(y)

plt.scatter(X_transed, y_transed)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_transed, y_transed, test_size=0.2, random_state=7)
model = LinearRegression().fit(X_train, y_train)
y_pred = model.predict(X_test)

print(f'Mean squared error = {MSE(y_test, y_pred)}')
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
```

-----------Хм. При StandardScaler преобразовании значение MSE такое же
