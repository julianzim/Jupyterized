```python
titanic = sns.load_dataset('titanic')
titanic = titanic.dropna(how='any', axis=0) #ÑƒÐ´Ð°Ð»Ð¸Ð» ÑÑ‚Ñ€Ð¾ÐºÐ¸ Ñ Ð¿ÑƒÑÑ‚Ñ‹Ð¼Ð¸ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸ÑÐ¼Ð¸
titanic
```

```python
%matplotlib inline

plt.scatter(titanic['age'], titanic['fare'])
```

```python
X, y = np.array(titanic['age'])[:, np.newaxis], np.array(titanic['fare'])[:, np.newaxis]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7) 
#ÑÐ¾Ð·Ð´Ð°Ð» Ñ‚Ñ€ÐµÐ½Ð¸Ñ€Ð¾Ð²Ð¾Ñ‡Ð½ÑƒÑŽ Ð¸ Ñ‚ÐµÑÑ‚Ð¾Ð²ÑƒÑŽ Ð²Ñ‹Ð±Ð¾Ñ€ÐºÐ¸
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()    #ÑÐ¾Ð·Ð´Ð°ÐµÐ¼ ÑÐºÐ·ÐµÐ¼Ð¿Ð»ÑÑ€ Ð¼Ð¾Ð´ÐµÐ»Ð¸
model.fit(X_train, y_train)   #Ð¾Ð±ÑƒÑ‡Ð°ÐµÐ¼ Ð¼Ð¾Ð´ÐµÐ»ÑŒ
```

```python
y_pred = model.predict(X_test)   #Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ñ‹Ð²Ð°ÐµÐ¼ Ñ‚Ð°Ñ€Ð³ÐµÑ‚ Ð½Ð° Ñ‚ÐµÑÑ‚Ð¾Ð²Ñ‹Ñ… Ñ„Ð¸Ñ‡Ð°Ñ…
```

```python
model.coef_   #Ð½Ð°Ð¹Ð´ÐµÐ½Ð½Ñ‹Ð¹ ÐºÐ¾ÑÑ„Ñ„Ð¸Ñ†Ð¸ÐµÐ½Ñ‚ Ð¼Ð¾Ð´ÐµÐ»Ð¸ Ð»Ð¸Ð½ÐµÐ¹Ð½Ð¾Ð¹ Ñ€ÐµÐ³Ñ€ÐµÑÑÐ¸Ð¸
```

```python
#Ð¿Ñ€Ð¾Ð²ÐµÑ€ÑÐµÐ¼ Ð¼ÐµÑ‚Ñ€Ð¸ÐºÐ¸ ÐºÐ°Ñ‡ÐµÑÑ‚Ð²Ð°

from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import r2_score

print(f'Mean squared error = {MSE(y_test, y_pred)}')
print(f'R2_Score = {r2_score(y_test, y_pred)}')
```

```python
#Ð²Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð¸Ñ€ÑƒÐµÐ¼ Ñ‚ÐµÑÑ‚Ð¾Ð²ÑƒÑŽ Ð²Ñ‹Ð±Ð¾Ñ€ÐºÑƒ Ð¸ Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð¸Ðµ

plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red', linewidth=3)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
titanic = sns.load_dataset('titanic').dropna(how='any', axis=0)
titanic
```

```python
X, y = np.array(titanic['age'])[:, np.newaxis], np.array(titanic['fare'])[:, np.newaxis]
```

```python
plt.scatter(X, y)   #ÑÑ‹Ñ€Ñ‹Ðµ Ð´Ð°Ð½Ð½Ñ‹Ðµ. ÐµÑÑ‚ÑŒ Ð²Ñ‹Ð±Ñ€Ð¾ÑÑ‹ Ð¿Ð¾ Ñ†ÐµÐ»ÐµÐ²Ð¾Ð¹ Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½Ð¾Ð¹
```

```python
y_trans = np.sqrt(y)   #ÐµÑÐ»Ð¸ Ð¿Ñ€Ð¾Ð»Ð¾Ð³Ð°Ñ€Ð¸Ñ„Ð¼Ð¸Ñ€Ð¾Ð²Ð°Ñ‚ÑŒ Ñ‚Ð¾ Ð¿Ð¾ÑÐ²ÑÑ‚ÑÑ Ð±ÐµÑÐºÐ¾Ð½ÐµÑ‡Ð½Ñ‹Ðµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ. Ð¿Ð¾ÑÑ‚Ð¾Ð¼Ñƒ Ð²Ð·ÑÐ» ÐºÐ¾Ñ€ÐµÐ½ÑŒ
plt.scatter(X, y_trans)   #Ð²Ñ‹Ð±Ñ€Ð¾ÑÑ‹ ÑÑ‚Ð°Ð»Ð¸ ÑƒÐ¶Ðµ Ð½Ð°Ð¼Ð½Ð¾Ð³Ð¾ Ð¼ÐµÐ½ÑŒÑˆÐµ
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as MSE

X_train, X_test, y_train, y_test = train_test_split(X, y_trans, test_size=0.2, random_state=7)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(f'Mean squared error = {MSE(y_test, y_pred)}')
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
```

```python
from sklearn.preprocessing import MinMaxScaler

X_transed = MinMaxScaler().fit_transform(X)
y_transed = MinMaxScaler().fit_transform(y)

plt.scatter(X_transed, y_transed)
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
```
