```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,10,1000)

fig = plt.figure()
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x), '-.g', label='Sin(x)')
ax.plot(x, np.cos(x), label='Cos(x)')
plt.title('Sin(x) and Cos(x)')
plt.xlabel('x')
plt.ylabel('Sin/Cos')
plt.legend()
plt.axis([-2,12,-2,2])
```

```python
peng = sns.load_dataset('penguins')
df = pd.DataFrame(peng)
print('всего значений пола - {}'.format(df['sex'].count()))
print('mean body mass - {}'.format(df['body_mass_g'].mean()))
print('median body mass - {}'.format(df['body_mass_g'].median()))
print('СКО body mass - {}'.format(df['body_mass_g'].std()))
print('дисперсия body mass - {}'.format(df['body_mass_g'].var()))
print('min and max - {}, {}'.format(df['body_mass_g'].min(), df['body_mass_g'].max()))
df.head(2)
```
